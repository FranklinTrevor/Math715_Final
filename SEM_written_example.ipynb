The numerical implementation of the SEM solver done in this work can be analytically verified below
The domain is from $0\le x \le 1$ and is made of 2 equally sized elements and 3 nodes. 
The polynomial order used for this example is 2 which corresponds to 3 Nodes in an element.

First, the basis functions are constructed on the reference element for $P=2$ using the known roots of the GLL points shown in the Table blow.

**For \( P = 2 \rightarrow N = 3 \)**

| \(\zeta_i\) | \(w_i\) |
|------------:|--------:|
| \(-1\)      | \(1/3\) |
| \(0\)       | \(4/3\) |
| \(1\)       | \(1/3\) |

$$
\phi_j(\zeta) = \prod^{N}_{i=0,j\ne i} \frac{\zeta-\zeta_i}{\zeta_j - \zeta_i}, \ \ \ i,j = 0,...,N
$$
$$
\phi_0(\zeta) = \left( \frac{\zeta-0}{-1 - 0} \right) \left( \frac{\zeta-1}{-1 - 1}\right)
$$
$$
\phi_1(\zeta) = \left( \frac{\zeta+1}{0 + 1} \right) \left( \frac{\zeta-1}{0 - 1}\right)
$$
$$
\phi_2(\zeta) = \left( \frac{\zeta+1}{1 + 1} \right) \left( \frac{\zeta-0}{1 - 0}\right)
$$
Plotting the basis functions across the reference element are shown below

```python
#Analytical 
zeta = np.linspace(-1,1,100)
bf_exact_0 = ((zeta-0)/(-1-0))*((zeta-1)/(-1-1))
bf_exact_1 = ((zeta+1)/(0+1))*((zeta-1)/(0-1))
bf_exact_2 = ((zeta+1)/(1+1))*((zeta-0)/(1-0))

#Code
weights, roots, zeta_c, bf = fn.Lagrange_basis_functions(2)

plt.figure(figsize = (12,8))
plt.plot(zeta_c, bf[0,:], linewidth = 4, color = '#1f77b4', label = r'$\phi_0$')
plt.plot(zeta_c, bf[1,:], linewidth = 4, color = '#ff7f0e', label = r'$\phi_1$')
plt.plot(zeta_c, bf[2,:], linewidth = 4, color = '#2ca02c', label = r'$\phi_2$')
plt.plot(zeta, bf_exact_0, linewidth = 2, color = 'black', linestyle = 'dashed', label = 'Exact')
plt.plot(zeta, bf_exact_1, linewidth = 2, color = 'black', linestyle = 'dashed')
plt.plot(zeta, bf_exact_2, linewidth = 2, color = 'black', linestyle = 'dashed')
plt.xlabel(r'$\zeta$', fontsize = 20)
plt.ylabel(r'$\phi_j$', fontsize = 20)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize = '16')
plt.xticks([-1,0,1], fontsize = '14')
plt.yticks([0,1],fontsize = '14')
plt.grid()
plt.tight_layout()
plt.show()
```
